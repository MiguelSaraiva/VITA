```python
import requests
import time
import pandas as pd

API_KEY = my_key
RADIUS = 5000  # search within 5000 meters
TYPE = 'food'  # type of place
SECTIONS = 10  # number of sections to divide the radius

restaurant_count_list = []

# Function to get the count of restaurants for a given location
def get_restaurant_count(location):
    lat, lon = location.split(',')
    count = 0
    restaurant_ids = set()

    # Calculate the section size
    section_radius = RADIUS // SECTIONS

    for i in range(SECTIONS):
        section_lat = float(lat) + (i * (section_radius / 111000))
        section_lon = float(lon) + (i * (section_radius / (111000 * abs(math.cos(math.radians(float(lat)))))))

        url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={section_lat},{section_lon}&radius={section_radius}&type={TYPE}&key={API_KEY}"

        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            results = response.json()

            # Increment the count by the number of unique establishments found in the current section
            for result in results['results']:
                restaurant_id = result['place_id']
                if restaurant_id not in restaurant_ids:
                    restaurant_ids.add(restaurant_id)
                    count += 1
        else:
            print(f"Request failed with status code {response.status_code}")
            break

        # Pause for 2 seconds before making the next section request
        time.sleep(2)

    restaurant_count_list.append(count)
    return count

# Initialize variables to keep track of municipality changes
prev_municipality = None
iteration = 1

# Iterate over the rows of the DataFrame
for index, row in df.iterrows():
    # Get the current municipality and zone coordinates
    municipality = row['zone_name']
    coordinates = row['zone_coords']

    # Print the municipality and iteration count when it changes
    if municipality != prev_municipality:
        print(f"\n{municipality} - Iteration: {iteration}")
        iteration += 1

    # Get the count of restaurants for the current location
    count = get_restaurant_count(coordinates)

    # Print the count for the current location
    print(f"Count for {coordinates}: {count}")

    # Pause for 5 seconds before proceeding to the next zone
    time.sleep(5)

    # Update the previous municipality variable
    prev_municipality = municipality
```
